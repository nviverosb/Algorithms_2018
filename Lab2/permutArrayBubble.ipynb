# Bubble sort over all permutations of array

```python
#Taken from http://disi.unal.edu.co/~algorithms/python/insertionsort.py
#First, let's import libraries
```

```python
import random
import matplotlib.pyplot as plt
import time
```

```python
#Define random permutations function
def randomPerm(n):
    v=[]
    for i in range(n):
        v.append(i+1)
    for i in range(len(v)-1):
        j = random.randint(i, len(v)-1)
        aux = v[i]
        v[i] = v[j]
        v[j] = aux
    return v
```

```python
#Define Insertion sort with step count function
```

```python
def bsortSteps(a):
    v = []
    for i in range(len(a)):
        v.append(a[i])

    steps = 0
    
    for i in range(len(v)-1):
        for j in range(len(v)-1):
            if v[j] > v[j+1]:
                temp = v[j]
                v[j] = v[j+1]
                v[j+1] = temp
                steps = steps + 3
            steps = steps + 1
    steps = steps + 1
    
    return steps
```

```python
#How many times will I run the code
n = 100
runs = 1000
```

```python
#Run this bad boy
sum = 0
min = n**3
max = 0
arr = []
for i in range(runs):
    #tim = time.time()
    t = bsortSteps(randomPerm(n))
    arr.append(t)
    print(t)
    if t < min :
       min = t
    if t > max :
       max = t
    sum = sum + t
```

```python
#Print the results
print ('Theoretical best time, ' , str(5*n - 4))
print ('Theoretical worst time,' , str(((n+1)*n)/2))
print ('Theoretical average time,' , str(((n+1)*n)/4))
print ('Experimenal best time, ' , str(min))
print ('Experimenal worst time,' , str(max))
print ('Experimenal average time,' , str(sum/runs))
```
