# Time vs Fibonacci number Graph

First, libraries need to be imported and fibonacci function has to be declared

```python
import random
import time
import numpy as np
```

```python
def fibo(n):
    if n==0 :
        return 0
    elif n==1:
        return 1
    else:
        return fibo(n-1) + fibo(n-2);
```

In order to plot the desired data, I will store the values of the fibonacci numbers into a single array, which is going to be plotted

```python
fiboArray = [0 for x in range(50)]

for k in range(len(fiboArray)):
    fiboArray[k] = fibo(k)
```

```python
import matplotlib.pyplot as plt
plt.plot(fiboArray, color='lightblue', linewidth=3)
#plt.xlim(0.5, 4.5)
plt.xlabel('Numbers')
plt.ylabel('Times')
plt.legend(['Times']);
plt.show()
```
