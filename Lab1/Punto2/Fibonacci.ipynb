# Time vs Fibonacci number Graph

First, libraries need to be imported and fibonacci function has to be declared

```python
import random
import time
import numpy as np
```

```python

def fibo(n):
    
    start = time.time()
    first,second = 0,1
    for i in range(n):
        first,second = second,first+second
    start = time.time() -start
    return start
    """if n==0 :
        return 0
    elif n==1:
        return 1
    else:
        return fibo(n-1) + fibo(n-2);"""
```

In order to plot the desired data, I will store the values of the fibonacci numbers into a single array, which is going to be plotted

```python
#print fibo(50)
fiboArray = [0 for x in range(10000)]

for k in range(len(fiboArray)):
    #print fibo(k)
    fiboArray[k] = fibo(k)
```

```python
import matplotlib.pyplot as plt
plt.plot(fiboArray, color='lightblue', linewidth=1,drawstyle='steps')
#plt.xlim(0.5, 4.5)
plt.xlabel('Numbers')
plt.ylabel('Times')
plt.legend(['Times']);
plt.show()
```
