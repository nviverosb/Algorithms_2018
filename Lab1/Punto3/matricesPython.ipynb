```python
import time
import matplotlib.pyplot as plt
import numpy
```

```python
#Matrix Function definition
def Matrix(A):

    start = time.time()
    numeroDeInstrucciones = 2*A*A*A-A*A
    MatrixA = [[1 for x in range(A)] for y in range(A)]
    MatrixB = [[2 for x in range(A)] for y in range(A)]
    result = [[0 for x in range(A)] for y in range(A)]

    for i in range(len(MatrixA)):
        # iterate through columns of Y
        for j in range(len(MatrixB[0])):
            # iterate through rows of Y
            for k in range(len(MatrixB)):
                result[i][j] = result[i][j] + MatrixA[i][k] * MatrixB[k][j]
    end = time.time()

    return((end-start)/numeroDeInstrucciones)
```

```python
#main function definition
def main():
    n = 100
    #print(fib(10))
    resultados = [[0 for x in range(n)] for y in range(10)]
    promedios = [0 for x in range(n)]

    for i in range(10):
        for j in range(n):
            resultados[i][j] = Matrix(j+1)

    for i in range(n):
        for j in range(10):
            promedios[i] = promedios[i] + resultados[j][i]/10

    #for i in range(n):
        #print(i, " ", promedios[i])
    
    return promedios
```

```python
A = main()
```

```python
plt.plot(A);
plt.xlabel('Matrix size')
plt.ylabel('Time per instruction [s]')
plt.legend(['python']);
```

```python
#using numpy
def MatrixNumpy(A):

    start = time.time()
    numeroDeInstrucciones = 2*A*A*A-A*A
    
    MatrizA = numpy.full((A,A),1)
    MatrizB = numpy.full((A,A),2)

    #result = numpy.zeros(shape=(A,A))
    result = MatrizA*MatrizB
    end = time.time()

    return((end-start)/numeroDeInstrucciones)
```

```python
def mainNumpy():
    n = 100
    #print(fib(10))
    resultados = numpy.full((10,n),0)
    promedios = numpy.full((n),1)

    for i in range(10):
        for j in range(n):
            resultados[i][j] = MatrixNumpy(j+1)

    for i in range(n):
        for j in range(10):
            promedios[i] = promedios[i] + resultados[j][i]/10

    #for i in range(n):
        #print(i, " ", promedios[i])
    
    return promedios
```

```python
B = mainNumpy()
```

```python
plt.plot(B);
plt.xlabel('Matrix Size') # The data we generated is unitless, but don't forget units in general.
plt.ylabel('Instruction time')
plt.legend(['numpy']);
```
